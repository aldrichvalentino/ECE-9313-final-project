```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import data1 and data2 csv
data1 = pd.read_csv('data1.csv')
data2 = pd.read_csv('data2.csv')

print(len(data1))
print(len(data2))
```

```python
data1.columns
```

```python
# create log of bidirectional_duration_ms
data1['log_bidirectional_duration_ms'] = np.log(data1['bidirectional_duration_ms']+1)
data2['log_bidirectional_duration_ms'] = np.log(data2['bidirectional_duration_ms']+1)

# visualize the distribution with labels in two graphs
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
# Set color yellow
sns.distplot(data1['log_bidirectional_duration_ms'], ax=ax[0], color="orange")
ax[0].set_title('Interface 1')
sns.distplot(data2['log_bidirectional_duration_ms'], ax=ax[1])
ax[1].set_title('Interface 2')
plt.show()
```

```python
# print the percentile distribution in increments of 5
ranges = np.arange(0.05, 1.05, 0.05)
data1['log_bidirectional_duration_ms'].describe(percentiles=ranges)
```
